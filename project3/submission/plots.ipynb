```python
with open('bbox_sizes.json', 'r') as f:
    bbox_sizes = json.load(f)

with open('all_times.json', 'r') as f:
    all_times = json.load(f)
```

```python
import collections
import numpy as np

grouped = collections.defaultdict(list)
for area, seq in bbox_sizes:
    grouped[seq].append(area)

mean_bbox_by_seq_dict = {np.mean(areas): seq  for seq, areas in grouped.items()}
mean_bbox_by_seq = [np.mean(areas) for areas in grouped.values()]
```

```python
mean_bbox_by_seq_dict
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

seq_names = list(mean_bbox_by_seq_dict.values())
mean_areas = list(mean_bbox_by_seq)

cmap = cm.get_cmap('tab20', len(seq_names))  
def moving_average(x, w=20):
    return np.convolve(x, np.ones(w)/w, mode='same')

plt.figure(figsize=(10, 6))

for i, (time, seq_name) in enumerate(zip(all_times, seq_names)):
    color = cmap(i)
    label = f"{seq_name} ({int(mean_bbox_by_seq[i])} px)"
    time = moving_average(time, 5) 
    plt.plot(np.arange(1, len(time) + 1), time, label=label, color=color)

plt.xlabel("Frame", fontsize=12)
plt.ylabel("Processing Time (s)", fontsize=12)
plt.legend(fontsize=10, loc='upper right', ncol=2)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import collections

grouped_areas = collections.defaultdict(list)
for area, seq in bbox_sizes:
    grouped_areas[seq].append(area)

mean_bbox_by_seq = {seq: np.mean(areas) for seq, areas in grouped_areas.items()}

all_seq_names = mean_bbox_by_seq.keys()

grouped_times = collections.defaultdict(list)
for time, seq in zip(all_times, all_seq_names):  
    grouped_times[seq].extend(time)


mean_time_by_seq = {seq: np.mean(times) for seq, times in grouped_times.items()}
for k, v in mean_time_by_seq.items():
    mean_time_by_seq[k] = 1 / mean_time_by_seq[k]

sorted_seqs = sorted(mean_time_by_seq, key=mean_time_by_seq.get, reverse=True)
mean_times_sorted = [mean_time_by_seq[seq] for seq in sorted_seqs]
mean_bboxes_sorted = [mean_bbox_by_seq[seq] for seq in sorted_seqs]

plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(sorted_seqs)), mean_times_sorted, color=cm.tab20(np.arange(len(sorted_seqs)) % 20))
plt.xticks(range(len(sorted_seqs)), sorted_seqs, rotation=45, ha='right', fontsize=9)
plt.ylabel("Mean FPS", fontsize=14)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

for i, (bar, bbox_size) in enumerate(zip(bars, mean_bboxes_sorted)):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{int(bbox_size)} px",
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
